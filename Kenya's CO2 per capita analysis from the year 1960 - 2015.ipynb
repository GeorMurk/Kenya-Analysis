```python
hist_indicator = 'CO2 emissions \(metric'
hist_country = 'KEN'

my_country1 = indicators['IndicatorName'].str.contains(hist_indicator)
my_country2 = indicators['CountryCode'].str.contains(hist_country)

kenya_data = indicators[my_country1 & my_country2]
```

```python
kenya_data
```

```python
plt.plot(kenya_data['Year'].values, kenya_data['Value'].values)

plt.xlabel('Years')
plt.ylabel('Emmissions')
plt.title('Kenya CO2 Emmission Omissions per Capita in Metric Tons')
plt.axis([1960, 2012, 0, 0.5])

plt.grid(True)

plt.show()
```
